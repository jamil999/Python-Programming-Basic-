0 1 1 2 3 5 8\
======================\
**Function Call:**\
fibonacci(5)\

0 1 1 2 3 5 

```python
# to do
number=int(input('Enter the limit of fibonacci series:'))

def fibonacci(number):
    first_number=0
    latest_number=1
    last_number=1
    print(first_number,end=' ')
    print(latest_number,end=' ')
    while(last_number<number):
        last_number=first_number+latest_number
        if last_number<=number:
            print(last_number,end=' ')
        first_number=latest_number
        latest_number=last_number
        
fibonacci(number)
```

For each player: pass the name, yearly earning, the total goal scored this season, bonus percent per goal. Additionally, if the goal scored is above 30, add an extra bonus of 10000 taka. If it is between 20 and 30 inclusive, add an extra 5000 taka.

**Example:**\
function_name("Neymar", 1200000, 30, 10)

bonus = 30 * (10 / 100 * 1200000) + 5000 = 3605000

**Function call:**\
function_name("Neymar", 1200000, 30, 10)\
**Output:**\
Neymar earned a bonus of 3605000 Taka for 30 goals.

=====================================================

**Function call:**\
function_name("Neymar", 1200000, 30, 10, "Jamal", 700000, 19, 5)\
**Output:**\
Neymar earned a bonus of 3605000 Taka for 30 goals.\
Jamal earned a bonus of 665000 Taka for 19 goals.





```python
# to do
def calculate_bonus(*player_info):
    count=0
    while count<len(player_info):
        if player_info[count+2]>30:
            bonus_earned=player_info[count+2]*(player_info[count+3]/100*player_info[count+1])+10000
        elif player_info[count+2]>=20 and player_info[count+2]<=30:
            bonus_earned=player_info[count+2]*(player_info[count+3]/100*player_info[count+1])+5000
        else:
            bonus_earned=player_info[count+2]*(player_info[count+3]/100*player_info[count+1])

        
        print(player_info[count+0],'earned a bonus of',int(bonus_earned),'Taka for',player_info[count+2],'goals.')
        count=count+4
        
calculate_bonus("Neymar",1200000,30,10,"Jamal",700000,19,5)
```
